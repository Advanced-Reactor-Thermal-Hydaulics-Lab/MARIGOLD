## ___DRIFT-FLUX ANALYSIS & RELATIVE VELOCITY___

```python
def assign_marker(cond, shapeset='jf'):
    if shapeset == 'jf':
        if abs(cond.jf - 5.0) < 0.001:
            cond.marker_type = 'o'
        elif abs(cond.jf - 3.8) < 0.001:
            cond.marker_type = 's'
        elif abs(cond.jf - 2.0) < 0.001:
            cond.marker_type = '^'
        
        if abs(cond.jgref - 0.23) < 0.001:
            cond.marker_color = 'red'
        elif abs(cond.jgref - 0.10) < 0.001:
            cond.marker_color = 'blue'

    elif shapeset == 'db':
        if cond.database == 'ryan':
            cond.marker_type = 'o'
            cond.marker_color = 'red'
        elif cond.database == 'adix':
            cond.marker_type = 's'
            cond.marker_color = 'lime'
        elif cond.database == 'neup':
            cond.marker_type = '^'
            cond.marker_color = 'blue'

    elif shapeset == 'tag':
        if cond.tag == 'DP4-2':
            # Parylene
            cond.marker_type = 'D'
            cond.marker_color = 'magenta'
        elif cond.tag == 'DM2-2' or cond.tag == 'SM2-1':
            # Two-sensor standard
            cond.marker_type = 'X'
            cond.marker_color = 'orange'
        elif cond.tag == 'LM4-10N' or cond.tag == 'LM4-11N':
            # NEUP repeatability tests
            cond.marker_type = 'v'
            cond.marker_color = 'cyan'
        elif cond.database == 'ryan':
            cond.marker_type = 'o'
            cond.marker_color = 'red'
        elif cond.database == 'adix':
            cond.marker_type = 's'
            cond.marker_color = 'lime'
        elif cond.database == 'neup':
            cond.marker_type = '^'
            cond.marker_color = 'blue'

def legend_decoys(shapeset='jf'):
    if shapeset == 'jf':
        plt.scatter([-10], [-10], linestyle='None', color='black', edgecolor='k', marker = '^', label = r'$\langle j_{f} \rangle = 2.0\ \ [m/s]$')
        plt.scatter([-10], [-10], linestyle='None', color='black', edgecolor='k', marker = 's', label = r'$\langle j_{f} \rangle = 3.8\ \ [m/s]$')
        plt.scatter([-10], [-10], linestyle='None', color='black', edgecolor='k', marker = 'o', label = r'$\langle j_{f} \rangle = 5.0\ \ [m/s]$')
        plt.scatter([-10], [-10], linestyle='None', color='blue', edgecolor='k', marker = 'o', label = r'$\langle j_{g} \rangle = 0.11\ \ [m/s]$')
        plt.scatter([-10], [-10], linestyle='None', color='red', edgecolor='k', marker = 'o', label = r'$\langle j_{g} \rangle = 0.23\ \ [m/s]$')

    elif shapeset == 'db':
        plt.scatter([-10], [-10], linestyle='None', color='red', edgecolor='k', marker = 'o', label = r'Ryan')
        plt.scatter([-10], [-10], linestyle='None', color='lime', edgecolor='k', marker = 's', label = r'Dix, Quan')
        plt.scatter([-10], [-10], linestyle='None', color='blue', edgecolor='k', marker = '^', label = r'NEUP')
    
    elif shapeset == 'tag':
        plt.scatter([-10], [-10], linestyle='None', color='red', edgecolor='k', marker = 'o', label = r'Ryan')
        plt.scatter([-10], [-10], linestyle='None', color='lime', edgecolor='k', marker = 's', label = r'Dix, Quan')
        plt.scatter([-10], [-10], linestyle='None', color='blue', edgecolor='k', marker = '^', label = r'NEUP')

        plt.scatter([-10], [-10], linestyle='None', color='magenta', edgecolor='k', marker = 'D', label = r'Parylene')
        plt.scatter([-10], [-10], linestyle='None', color='orange', edgecolor='k', marker = 'X', label = r'Two-Sensor Standard')
        plt.scatter([-10], [-10], linestyle='None', color='cyan', edgecolor='k', marker = 'v', label = r'NEUP Retakes')

def origin_line(x, m):
    return m * x

def dfm_analysis(database,mode,shapeset='jf',method='legacy'):
    plt.rcParams.update({'font.size': 12})
    plt.rcParams["font.family"] = "Times New Roman"
    plt.rcParams["mathtext.fontset"] = "cm"
    plt.figure()

    xs = []
    ys = []

    for cond in database:
        cond:Condition
        # print(cond)

        if mode == 'vgj':
            # Global measurements
            x = cond.j
            # y = cond.jgloc / mg.area_avg(cond,'alpha',method=method)
            # y = mg.area_avg(cond,'alpha_ug',method=method) / mg.area_avg(cond,'alpha',method=method)
            y = mg.void_area_avg(cond,'ug1',method=method)

            # Local measurements
            # x = mg.area_avg(cond,'jf',method=method)+mg.area_avg(cond,'alpha_ug',method=method)
            # y = mg.void_area_avg(cond,'ug1',method=method)

        elif mode == 'alphabeta':
            x = cond.jgloc / cond.j
            y = mg.area_avg(cond,'alpha',method=method)

        mpbl = plt.scatter(x, y, marker=cond.marker_type, color = cond.marker_color, edgecolor='k')

        xs.append(x)
        ys.append(y)

    xs = np.asarray(xs)
    ys = np.asarray(ys)
    x_fit = np.arange(10)

    if mode == 'vgj':
        m, b, r, _, _ = linregress(xs, ys)
        r2 = r**2
        textstr = "\n".join([r"$\langle \langle v_{g} \rangle \rangle $ = %0.3f$\langle j \rangle$ + %0.3f " %(m, b), r"$R^{2}$=%0.3f" %r2])
        ub = 7
        
        plt.title(f"${int(cond.theta)}^o$ Drift Flux Analysis")
        plt.xlabel(r"$\langle j \rangle \ [m/s]$")
        plt.ylabel(r"$\langle \langle v_{g} \rangle \rangle \ [m/s]$")
        
    elif mode == 'alphabeta':
        popt, _ = curve_fit(origin_line, xs, ys)
        m = popt[0]
        b = 0
        textstr = r"$\langle \alpha \rangle $ = %0.3f$\langle \beta \rangle$ " %m
        ub = 0.1

        plt.plot(x_fit, x_fit, marker='None', linestyle='--', color='black', alpha=0.3)
        plt.title(f"${int(cond.theta)}^o$ Alpha-Beta Analysis")
        plt.xlabel(r"$\langle \beta \rangle \ [-]$")
        plt.ylabel(r"$\langle \alpha \rangle \ [-]$")
        
    legend_decoys(shapeset=shapeset)
    
    y_fit = m * x_fit + b
    props = dict(boxstyle='round', facecolor='white', alpha=0)

    plt.plot(x_fit, y_fit, marker='None', linestyle='-', color='black')
    plt.text(0.8*ub, 0.1*ub, textstr, fontsize=12, verticalalignment='top', bbox=props, ha='center', va='center')
    plt.xlim(0, ub)
    plt.ylim(0, ub)
    plt.legend()
    plt.show()

    return m, b
```

```python
passkey = "cond.LoverD < 120 and cond.jf >= 2.0"
shapeset = 'db'
method = 'trapezoid'

database = neup_database + ryan_database + y_database
```

```python
dfm_database    = []

vgjs_loc        = []
vgjs_glo        = []
aa_vrs          = []
vw_vrs          = []
thetas          = []

for cond in database:
    if eval(passkey):
        dfm_database.append(cond)
        
dfm_database.sort(key=lambda cond: cond.theta, reverse=True)
print(f"theta\tjf\tjgP1\tport\tbase\tprobe\t\t<alpha>\t<<vg1>>\t<<Vgj>>")
print(f"---------------------------------------------------------------------------------------")
for cond in dfm_database:
    
    if cond.theta == 90:
        cond.mirror(method='axisym',force_remirror=True)
    else:
        cond.mirror(method='sym90')
        
    mg.approx_vf(cond)
    mg.calc_vgj(cond,method='approx',verbose=False)
    mg.calc_diff(cond,'ug1','vf_approx')

    vg          = mg.void_area_avg(cond,'ug1',method=method)
    j           = cond.j            # mg.area_avg(cond,'j',method=method)
    c0          = mg.area_avg(cond,'alpha_j',method=method) / ( mg.area_avg(cond,'alpha',method=method) * mg.area_avg(cond,'j',method=method))
    vgj_loc     = mg.void_area_avg(cond,'vgj',method=method)
    vgj_glo     = vg - (c0 * j)
    aa_vr       = mg.area_avg(cond,'diff')
    vw_vr       = mg.void_area_avg(cond,'diff')
    
    vgjs_loc.append(vgj_loc)        # Vgj, calculated using local variables
    vgjs_glo.append(vgj_glo)        # Vgj, calculated using global variables
    aa_vrs.append(aa_vr)            # Area-averaged vr
    vw_vrs.append(vw_vr)            # Void-weighted area-averaged vr
    thetas.append(cond.theta)       # Pipe inclination angle

    print(f"{cond.theta}\t{cond.jf}\t{cond.jgref}\t{cond.port}\t{cond.database}\t{cond.tag}\t\t{mg.area_avg(cond,'alpha'):0.3f}\t{vg:0.3f}\t{vgj_glo:0.3f}\t{c0:0.3f}")
    assign_marker(cond,shapeset=shapeset)

vgjs_loc        = np.array(vgjs_loc)
vgjs_glo        = np.array(vgjs_glo)
aa_vrs          = np.array(aa_vrs)
vw_vrs          = np.array(vw_vrs)
theta_set       = sorted(list(set(thetas)), reverse=True)
```

___VG-J___

```python
c0_fits     = []
vgj_fits    = []

for theta in theta_set:
    temp_database = []

    for cond in dfm_database:
        if cond.theta == theta:
            temp_database.append(cond)
    
    m, b = dfm_analysis(temp_database,'vgj',shapeset=shapeset)

    c0_fits.append(m)
    vgj_fits.append(b)
```

```python
k_fits      = []

for theta in theta_set:
    temp_database = []

    for cond in dfm_database:
        if cond.theta == theta:
            temp_database.append(cond)
    
    m, b = dfm_analysis(temp_database,'alphabeta',shapeset=shapeset)

    k_fits.append(m)
```

___PLOT PARAM1 VS. PARAM2___

```python
def plot_params(obj, param1, param2, include_complement = True, const_to_plot = [90, 67.5, 45, 22.5, 0], fig_size = None, xlim = None, ylim = None, xlabel = None, ylabel = None, title = None, legend_loc = None, auto_marker = True, logopt = False, suppress_legend = False):
    plt.rcParams.update({'font.size': 10})
    plt.rcParams["font.family"] = "Times New Roman"
    plt.rcParams["mathtext.fontset"] = "cm"
    fig, ax = plt.subplots(figsize=fig_size, dpi=300, layout='constrained')
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    ax.tick_params(direction='in',which='both')
    ms = mg.marker_cycle()
    cs = mg.color_cycle()

    if not hasattr(obj,'__len__'):
        cond = obj

        if cond.theta == 90:
            cond.mirror('axisym')
        else:
            cond.mirror('sym90')

        for angle in const_to_plot:
            r_dict = cond.data[angle]
            xparams = []
            yparams = []
            for r, midas_output in r_dict.items():
                try:
                    if midas_output[param2] != 0:
                        xparams.append(midas_output[param1])
                        yparams.append(midas_output[param2])
                except:
                    pass
            if include_complement:
                if angle > 180:
                    print("Error: Cannot find complement to angle > 180. Skipping")
                else:
                    r_dict = cond.data[angle+180]
                    for r, midas_output in r_dict.items():
                        try:
                            if midas_output[param2] != 0:
                                xparams.append(midas_output[param1])
                                yparams.append(midas_output[param2])
                        except:
                            pass
            
            ax.plot(xparams, yparams, label=f'{angle}°', color=next(cs), marker=next(ms), linestyle = '--')
    else:
        database = obj

        for cond in database:
            if cond.theta == 90:
                cond.mirror('axisym')
            else:
                cond.mirror('sym90')
                
            r_dict = cond.data[90]
            xparams = []
            yparams = []
            for r, midas_output in r_dict.items():
                try:
                    if midas_output[param2] != 0:
                        xparams.append(midas_output[param1])
                        yparams.append(midas_output[param2])
                except:
                    pass
            if include_complement:
                r_dict = cond.data[90+180]
                for r, midas_output in r_dict.items():
                    try:
                        if midas_output[param2] != 0:
                            xparams.append(midas_output[param1])
                            yparams.append(midas_output[param2])
                    except:
                        pass
            
            xparams, yparams = list(zip(*sorted(zip(xparams, yparams))))

            if auto_marker == True:
                ax.plot(xparams, yparams, label=f'{cond.theta}°, jf = {cond.jf}, jg = {cond.jgref}, {cond.port}', color=next(cs), marker=next(ms), linestyle = '--')
            else:
                assign_marker(cond)
                ax.plot(xparams, yparams, label=None, color=cond.marker_color, marker=cond.marker_type, linestyle = '--')

    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    
    if logopt:
        plt.semilogx()
    
    if not suppress_legend:
        if legend_loc == 'q1':
            plt.legend(loc = 'upper right')
        elif legend_loc == 'q2':
            plt.legend(loc = 'upper left')
        elif legend_loc == 'q3':
            plt.legend(loc = 'lower left')
        elif legend_loc == 'q4':
            plt.legend(loc = 'lower right')
        else:
            plt.legend(loc = 'upper left')

    if auto_marker == False:
        plt.scatter([-10], [-10], linestyle='None', color='deeppink', marker = 'o', label = r'$\theta = 90\degree$')
        plt.scatter([-10], [-10], linestyle='None', color='orange', marker = 's', label = r'$\theta = 80\degree$')
        plt.scatter([-10], [-10], linestyle='None', color='lime', marker = 'D', label = r'$\theta = 60\degree$')
        plt.scatter([-10], [-10], linestyle='None', color='cyan', marker = '^', label = r'$\theta = 30\degree$')
        plt.scatter([-10], [-10], linestyle='None', color='violet', marker = 'v', label = r'$\theta = 00\degree$')

        plt.legend()
        
    plt.axvline(0.0,color='k',linewidth=1)
    plt.axhline(0.0,color='k',linewidth=1)
    plt.show()
    
    return

def assign_marker(cond):
    if cond.theta == 90:
        cond.marker_color = 'deeppink'
        cond.marker_type = 'o'
    elif cond.theta == 80:
        cond.marker_color = 'orange'
        cond.marker_type = 's'
    elif cond.theta == 60:
        cond.marker_color = 'lime'
        cond.marker_type = 'D'
    elif cond.theta == 30:
        cond.marker_color = 'cyan'
        cond.marker_type = '^'
    elif cond.theta == 0:
        cond.marker_color = 'violet'
        cond.marker_type = 'v'
```
