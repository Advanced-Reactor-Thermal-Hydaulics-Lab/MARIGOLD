```python
def plot_signal(xs, ys, type):
    plt.figure(figsize=(6,5),constrained_layout=True)
    plt.plot(xs, ys, linestyle = 'solid', linewidth = 1, marker = '.')       #marker=None
    # plt.axhline(-0.03,color='k',linestyle='-')
    plt.axhline(0.0,color='k',linestyle='-')
    plt.legend(['sensor 0', 'sensor 1', 'sensor 2', 'sensor 3'],loc='upper right')
    plt.xlim([tmin, tmax])
    plt.ylim([vmin, vmax])

    if type == 'raw':
        plt.title(f'Raw Signals, dt = {dt}, tmin = {tmin:0.3f}')
    elif type == 'mdn':
        plt.title(f'Median Signals, dt = {dt}, tmin = {tmin:0.3f}')
    elif type == 'nrm':
        plt.title(f'Normalized Signals, dt = {dt}, tmin = {tmin:0.3f}')
    elif type == 'sqr':
        plt.title(f'Squared Signals, dt = {dt}, tmin = {tmin:0.3f}')

    plt.xlabel('Time [s]')
    plt.ylabel('Voltage [V]')
    plt.show()
```

## ___SIGNAL___

```python
thrs = [-0.05, -0.045, -0.04, -0.035, -0.03, -0.025, -0.02, -0.015, -0.01]

def longest_run_below(signal, thr):
    below = signal < thr
    longest = 0
    run = 0
    for b in below:
        if b:
            run += 1
            if run > longest:
                longest = run
        else:
            run = 0
    return longest

for thr in thrs:
    print(f"Threshold: {thr}")
    for idx, directory in enumerate(directories):
        nd = directories[idx] + paths[idx]
        nrm = np.loadtxt(os.path.join(nd, prefix + "_NormSig.dat"))

        runs = [longest_run_below(nrm[:, i], thr) for i in range(4)]
        print(os.path.basename(directory), *runs, sep="\t")
    print("\n")
```
